## ⚙️ Paso 2: Configuración del entorno en Google Colab

Una vez generado el archivo de autenticación `kaggle.json`, es necesario configurarlo en el entorno de Google Colab para poder utilizar la API de Kaggle.

Este proceso permite autenticar al entorno y habilitar la descarga automática de datasets.

### 📋 Instrucciones

1. Subir el archivo `kaggle.json` al entorno de Colab.

2. Crear la carpeta de configuración de Kaggle.

3. Mover el archivo a la ruta correspondiente.

4. Asignar permisos de seguridad adecuados.


```python
# Instalación de la librería Kaggle
!pip install kaggle

# Subir archivo kaggle.json
from google.colab import files
files.upload()
```

```python
# Configuración de credenciales

import os

# Crear carpeta de configuración
os.makedirs('/root/.kaggle', exist_ok=True)

# Mover archivo
os.rename('kaggle.json', '/root/.kaggle/kaggle.json')

# Asignar permisos (requerido por Kaggle)
os.chmod('/root/.kaggle/kaggle.json', 600)
```

### 🔐 Consideraciones de seguridad

El archivo `kaggle.json` contiene credenciales sensibles, por lo que:

- No debe ser compartido ni subido a repositorios públicos.
- Se utiliza únicamente dentro del entorno de ejecución.
- Los permisos asignados (`600`) restringen el acceso solo al usuario actual.

Este tipo de prácticas es fundamental en entornos reales de trabajo con datos.

### ⚠️ Intento inicial de descarga y error detectado

En el primer intento de descarga del dataset mediante la API de Kaggle, se obtuvo un error `403 Forbidden`.

Esto ocurre porque, en el caso de datasets de competencias, Kaggle requiere que el usuario acepte previamente los términos de participación desde la web mediante la opción **"Join Competition"**.

A continuación se muestra el intento inicial y luego la corrección aplicada.

```python
!kaggle competitions download -c house-prices-advanced-regression-techniques
```

### ✅ Corrección aplicada

Luego de aceptar manualmente los términos de la competencia en Kaggle, se volvió a ejecutar el comando de descarga y el dataset se descargó correctamente.
